***Exploratory Data Analysis (EDA)***

```python
def explore_data(df):
    
    print("=" * 55)
    print("          DATASET OVERVIEW")
    print("=" * 55)
    
    print(f"\n📊 Total Rows    : {df.shape[0]:,}")
    print(f"📊 Total Columns : {df.shape[1]}")
    
    print(f"\n📋 Column Data Types:")
    print("-" * 40)
    print(df.dtypes.to_string())
    
    print(f"\n⚠️  Missing Values Per Column:")
    print("-" * 40)
    missing = df.isnull().sum()
    missing_pct = (missing / len(df) * 100).round(2)
    missing_report = pd.DataFrame({
        'Missing Count' : missing,
        'Missing %'     : missing_pct
    })
    missing_report = missing_report[missing_report['Missing Count'] > 0]
    missing_report = missing_report.sort_values('Missing %', ascending=False)
    print(missing_report.to_string())
    
   
    print(f"\n🎯 Target Column (status) Balance:")
    print("-" * 40)
    counts = df['Status'].value_counts()
    pct    = df['Status'].value_counts(normalize=True) * 100
    print(f"  No Default (0) : {counts[0]:,}  ({pct[0]:.1f}%)")
    print(f"  Default    (1) : {counts[1]:,}  ({pct[1]:.1f}%)")
    
   
    print(f"\n🔁 Duplicate Rows : {df.duplicated().sum()}")
    print("=" * 55)

explore_data(df)
```

***Standardise Column Names***

def clean_column_names(df):
    
    original_cols = list(df.columns)
    
    df.columns = (
        df.columns
          .str.strip()           
          .str.lower()           
          .str.replace(' ', '_') 
          .str.replace('-', '_') 
    )
    
    print("✅ Column names cleaned!")
    print("\n📋 Before → After:")
    print("-" * 45)
    for old, new in zip(original_cols, df.columns):
        if old != new:
            print(f"  {old:35} → {new}")
    
    return df

df = clean_column_names(df)

***Fix Missing Values***

```python
def fix_missing_values(df):
    
    print("=" * 55)
    print("         FIXING MISSING VALUES")
    print("=" * 55)
    
    numeric_cols = [
        'rate_of_interest',
        'interest_rate_spread',
        'upfront_charges',
        'dtir1',
        'property_value',
        'ltv',
        'income',
        'term'
    ]
    
    print("\n📊 Numeric Columns → Filled with Median:")
    print("-" * 45)
    for col in numeric_cols:
        missing_before = df[col].isnull().sum()
        median_val     = df[col].median()
        df[col]        = df[col].fillna(median_val)
        print(f"  {col:30} | missing: {missing_before:,} | filled with: {median_val:.2f}")
    
    categorical_cols = [
        'loan_limit',
        'approv_in_adv',
        'loan_purpose',
        'neg_ammortization',
        'age',
        'submission_of_application'
    ]
    
    print("\n📋 Categorical Columns → Filled with Mode:")
    print("-" * 45)
    for col in categorical_cols:
        missing_before = df[col].isnull().sum()
        mode_val       = df[col].mode()[0]
        df[col]        = df[col].fillna(mode_val)
        print(f"  {col:30} | missing: {missing_before:,} | filled with: '{mode_val}'")
    
    remaining = df.isnull().sum().sum()
    print(f"\n✅ Done! Remaining nulls in dataset: {remaining}")
    
    return df

df = fix_missing_values(df)
```

***Feature Engineering: Age Category***

```python
def create_age_category(df):
    """
    Creates a new column 'age_category' from existing 'age' column.
    Maps age ranges to meaningful life-stage categories.
    This helps analyse default risk by life stage.
    """
    
    age_mapping = {
        '<25'   : '1. Young Adult',
        '25-34' : '2. Early Career',
        '35-44' : '3. Mid Career',
        '45-54' : '4. Peak Career',
        '55-64' : '5. Pre Retirement',
        '65-74' : '6. Senior',
        '>74'   : '7. Elderly'
    }
    
    df['age_category'] = df['age'].map(age_mapping)
    
    print("=" * 60)
    print("       AGE CATEGORY — DISTRIBUTION & DEFAULT RATE")
    print("=" * 60)
    print(f"\n{'Category':<22} {'Count':>8} {'Defaults':>10} {'Default Rate':>14}")
    print("-" * 60)
    
    for category in sorted(df['age_category'].unique()):
        subset       = df[df['age_category'] == category]
        total        = len(subset)
        defaults     = subset['status'].sum()
        default_rate = (defaults / total * 100)
        print(f"  {category:<20} {total:>8,} {defaults:>10,} {default_rate:>13.1f}%")
    
    print("-" * 60)
    print(f"  {'TOTAL':<20} {len(df):>8,} {df['status'].sum():>10,} {df['status'].mean()*100:>13.1f}%")
    print(f"\n✅ New column 'age_category' created!")
    print(f"   Total columns now: {df.shape[1]}")
    
    return df

df = create_age_category(df)
```

***Replace All Coded Values With Meaningful Labels***


```python
def replace_coded_values(df):
    
    print("=" * 60)
    print("       REPLACING CODED VALUES WITH LABELS")
    print("=" * 60)
    
    all_mappings = {
        
        'loan_purpose' : {
            'p3' : 'Home Purchase',
            'p4' : 'Commercial Purpose',
            'p1' : 'Refinancing',
            'p2' : 'Home Improvement'},
        
        'credit_worthiness' : {
            'l1' : 'Standard',
            'l2' : 'Sub-Standard'},
        
        'open_credit' : {
            'nopc' : 'No Open Credit',
            'opc'  : 'Has Open Credit'},
        
        'loan_type' : {
            'type1' : 'Conventional',
            'type2' : 'FHA Insured',
            'type3' : 'VA Loan'},
        
        'loan_limit' : {
            'cf'  : 'Conforming',
            'ncf' : 'Non-Conforming'},
        
        'approv_in_adv' : {
            'nopre' : 'Not Pre-Approved',
            'pre'   : 'Pre-Approved'},
        
        'neg_ammortization' : {
            'not_neg'  : 'Standard Amortization',
            'neg_amm'  : 'Negative Amortization'},
        
        'interest_only' : {
            'not_int'  : 'Principal + Interest',
            'int_only' : 'Interest Only'},
        
        'lump_sum_payment' : {
            'not_lpsm' : 'No Lump Sum',
            'lpsm'     : 'Lump Sum Payment'},
        
        'construction_type' : {
            'sb' : 'Site Built',
            'mh' : 'Manufactured Home'},
        
        'occupancy_type' : {
            'pr' : 'Primary Residence',
            'ir' : 'Investment Property',
            'sr' : 'Secondary Residence'},
        
        'total_units' : {
            '1u' : 'Single Unit',
            '2u' : 'Duplex',
            '3u' : 'Triplex',
            '4u' : 'Quadplex'}}
    
   
    print(f"\n{'Column':<25} {'Unique Before':<18} {'Unique After'}")
    print("-" * 60)
    
    for col, mapping in all_mappings.items():
        before = df[col].unique().tolist()
        df[col] = df[col].map(mapping)
        after  = df[col].unique().tolist()
        print(f"  {col:<23} {str(before):<18}")
        print(f"  {'':<23} → {str(after)}")
        print()
    
    print("-" * 60)
    print(f"✅ All coded values replaced!")
    print(f"   Remaining nulls: {df.isnull().sum().sum()}")
    
    return df


df = replace_coded_values(df)
```

***Fix total_units mapping***


```python
def fix_total_units(df):
    
    print("Current values in total_units:")
    print(df['total_units'].value_counts(dropna=False))
    
    df_raw = pd.read_csv(r"C:\Users\Nkita\Downloads\Loan_Default.csv")
    df_raw.columns = (df_raw.columns
                      .str.strip()
                      .str.lower()
                      .str.replace(' ', '_')
                      .str.replace('-', '_'))
    
    df['total_units'] = df_raw['total_units'].str.strip().str.upper()
    
    units_mapping = {
        '1U' : 'Single Unit',
        '2U' : 'Duplex',
        '3U' : 'Triplex',
        '4U' : 'Quadplex'
    }
    
    df['total_units'] = df['total_units'].map(units_mapping)
    
    print("\n✅ After fix:")
    print(df['total_units'].value_counts(dropna=False))
    print(f"\nNulls remaining: {df['total_units'].isnull().sum()}")
    
    return df
df = fix_total_units(df)
```

***Feature Engineering***

```python
print("=" * 60)
print("          FEATURE ENGINEERING")
print("=" * 60)

bins   = [0, 580, 670, 740, 800, 900]
labels = ['Poor', 'Fair', 'Good', 'Very Good', 'Exceptional']

df['credit_score_category'] = pd.cut(
    df['credit_score'],
    bins   = bins,
    labels = labels,
    right  = True
)

print("\n1️⃣  CREDIT SCORE CATEGORY:")
print("-" * 45)
for cat in labels:
    subset  = df[df['credit_score_category'] == cat]
    total   = len(subset)
    default = subset['status'].sum()
    rate    = (default / total * 100) if total > 0 else 0
    print(f"  {cat:<15} | Count: {total:>7,} | Default Rate: {rate:.1f}%")
```

***Ltv risk Category***

```python
bins   = [0, 60, 80, 90, 100, float('inf')]
labels = ['Low Risk', 'Moderate Risk', 'High Risk', 
          'Very High Risk', 'Extreme Risk']

df['ltv_risk_category'] = pd.cut(
    df['ltv'],
    bins   = bins,
    labels = labels,
    right  = True)

print("\n2️⃣  LTV RISK CATEGORY:")
print("-" * 45)
for cat in labels:
    subset  = df[df['ltv_risk_category'] == cat]
    total   = len(subset)
    default = subset['status'].sum()
    rate    = (default / total * 100) if total > 0 else 0
    print(f"  {cat:<18} | Count: {total:>7,} | Default Rate: {rate:.1f}%")
```

***Loan to income ratio***

```python
df['loan_to_income'] = (
    df['loan_amount'] / df['income'].replace(0, np.nan)
).round(2)

df['loan_to_income'] = df['loan_to_income'].fillna(
    df['loan_to_income'].median()
)

print("\n3️⃣  LOAN TO INCOME RATIO:")
print("-" * 45)
print(f"  Minimum  : {df['loan_to_income'].min():.2f}")
print(f"  Median   : {df['loan_to_income'].median():.2f}")
print(f"  Maximum  : {df['loan_to_income'].max():.2f}")
print(f"  Meaning  : Average person borrows "
      f"{df['loan_to_income'].median():.1f}x their annual income")
```

***loan amount category***

```python
bins   = [0, 150000, 300000, 500000, float('inf')]
labels = ['Small', 'Medium', 'Large', 'Jumbo']

df['loan_amount_category'] = pd.cut(
    df['loan_amount'],
    bins   = bins,
    labels = labels,
    right  = True
)

print("\n4️⃣  LOAN AMOUNT CATEGORY:")
print("-" * 45)
for cat in labels:
    subset  = df[df['loan_amount_category'] == cat]
    total   = len(subset)
    default = subset['status'].sum()
    rate    = (default / total * 100) if total > 0 else 0
    avg     = subset['loan_amount'].mean()
    print(f"  {cat:<8} | Count: {total:>7,} | "
          f"Avg: {avg:>10,.0f} | Default Rate: {rate:.1f}%")
```

***Fix Outliers Using IQR Method***

```python
print("=" * 60)
print("          FIXING OUTLIERS — IQR METHOD")
print("=" * 60)

numeric_cols = [
    'loan_amount',
    'property_value', 
    'income',
    'upfront_charges',
    'ltv',
    'dtir1',
    'rate_of_interest',
    'interest_rate_spread',
    'loan_to_income']

print(f"\n{'Column':<25} {'Before Min':>12} {'Before Max':>14} "
      f"{'After Min':>12} {'After Max':>12} {'Capped':>8}")
print("-" * 88)

for col in numeric_cols:
    
    Q1  = df[col].quantile(0.25)   # 25th percentile
    Q3  = df[col].quantile(0.75)   # 75th percentile
    IQR = Q3 - Q1                  # Interquartile range
    
    lower_bound = Q1 - (3 * IQR)   # anything below this = outlier
    upper_bound = Q3 + (3 * IQR)   # anything above this = outlier
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    
    before_min = df[col].min()
    before_max = df[col].max()
    
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    
    after_min = df[col].min()
    after_max = df[col].max()
    
    print(f"  {col:<23} {before_min:>12,.0f} {before_max:>14,.0f} "
          f"{after_min:>12,.0f} {after_max:>12,.0f} {outliers:>8,}")

print("-" * 88)
print(f"\n✅ Outliers capped successfully!")
print(f"   Method : IQR × 3 (conservative capping)")
print(f"   Nulls  : {df.isnull().sum().sum()}")
```
